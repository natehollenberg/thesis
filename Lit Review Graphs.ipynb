```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('basketball_reference_stats.csv')
```

```python
data
```

```python
plt.figure(figsize=(10,7))
plt.plot(data['YR'], data['threeptrt'], color = 'b')
plt.title("NBA Three Point Rate from 1979-2019", fontsize = 'large', fontweight ='bold')
plt.annotate('NBA Shortned 3-PT \n Line to Uniform 22 Feet', xy=(1996, 22), xytext=(1996, 28),
            arrowprops=dict(facecolor='black'), fontweight = 'bold')
plt.annotate('3% 3-PT Rate in 1980', xy=(1980, 4), xytext=(1980, 9),
            arrowprops=dict(facecolor='black'), fontweight = 'bold')
plt.annotate('35.9% 3-PT Rate Today', xy=(2019, 36), xytext=(2005, 35),
            arrowprops=dict(facecolor='black'), fontweight = 'bold')
plt.grid(False)
plt.ylabel('3-PT Rate in %', fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.savefig('threeptrate.png')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10,10))
ax1.plot(data['YR'], data['3PA'], color = 'b')
ax1.set_title("3-PT Attempts from 1979-2019", fontsize = 'large', fontweight = 'bold')
ax1.set_ylabel('3-PT Attempts', fontweight = 'bold')
ax1.set_xlabel('Year', fontweight = 'bold')
ax1.grid(False)
ax2.plot(data['YR'], data['2PA'], color = 'b')
ax2.set_title("2-PT Attempts from 1979-2019", fontsize = 'large', fontweight = 'bold')
ax2.set_ylabel('2-PT Attempts', fontweight = 'bold')
ax2.set_xlabel('Year', fontweight = 'bold')
ax2.grid(False)
fig.subplots_adjust(hspace=0.3)
plt.savefig('threeptattempts.png')
plt.show()
```

```python
plt.plot(data['YR'], data['3P%'], color = 'b')
plt.title("NBA 3-PT Percentage from 1979-2019", fontsize = 'large', fontweight ='bold')
plt.ylabel('3-PT %', fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.grid(False)
plt.savefig('threeptpct.png')
plt.show()
```

```python

plt.plot(wnba['Rk'], wnba['threeptrate'], color = 'b')
plt.grid(False)
plt.title("WNBA 3-PT Rate from 2001-2019", fontsize = 'large', fontweight ='bold')
plt.ylabel('3-PT Rate in %', fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.xticks(wnba['Rk'], rotation = 45)
plt.annotate('2013: Line Moved Back', xy=(2013, 21.6), xytext=(2011, 28),
            arrowprops=dict(facecolor='black'), fontweight = 'bold')
plt.annotate('2004: Line Moved Back', xy=(2004, 21), xytext=(2002, 28),
            arrowprops=dict(facecolor='black'), fontweight = 'bold')
plt.savefig('wnba_rate.png')
plt.show()
```

```python
plt.plot(wnba['Rk'], wnba['3pt%'], color = 'b')
plt.grid(False)
plt.title("WNBA 3-PT Percentage from 2001-2019", fontsize = 'large', fontweight ='bold')
plt.ylabel('3-PT Percentage', fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.xticks(wnba['Rk'], rotation = 45)
plt.savefig('wnba_percentage.png')
plt.show()
```
